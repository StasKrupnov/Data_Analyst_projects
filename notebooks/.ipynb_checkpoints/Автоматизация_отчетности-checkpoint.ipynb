# Цель

Создать систему оповещения о показателях избранных метрик социальной сети.  
Оповещения должны приходить каждый день в Telegram.  
Создадим telegram бота и автоматизируем отправку через Airflow. 

# Результат

# Код

```python
import telegram
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
import pandas as pd

from airflow.decorators import dag, task
from datetime import datetime, date, timedelta
import pandahouse as ph
import typing

default_args = {
    'owner': 's.krupnov',
    'depends_on_past': False,
    'retries': 3,
    'retry_delay': timedelta(minutes=5),
    'start_date': datetime(2023, 3, 28)
}


@dag(default_args=default_args, schedule_interval='0 11 * * *', catchup=False)
def dag_report_krupnov(None) -> None:
    '''
    Функция, вызывающая выполнение дага
    '''
    @task
    def extract(None) -> pd.DataFrame:
        '''
        Выгружает нужные данные из ClickHouse
        ----------
        Возвращает:
        df - датафрейм с нужными данными
        '''

        connection = {
            'host': 'https://clickhouse.lab.karpov.courses/',
            'password': 'dpo_python_2020',
            'user': 'student',
            'database': 'simulator_20230220'
        }

        q_feed = '''
                SELECT toDate(time) as event_date, user_id,
                sum(action = 'like') as likes,
                sum(action = 'view') as views
                FROM {db}.feed_actions
                WHERE toDate(time) >= today() - 6
                GROUP BY event_date, user_id
                order by event_date
                '''

        df = ph.read_clickhouse(query=q_feed, connection=connection)

        return df

    @task
    def transform_1(df: pd.DataFrame) -> str:
        '''
        Считает метрики ленты новостей за вчера
        ----------
        Аргументы:
        df - извлеченный из CH датафрейм
        ----------
        Возвращает:
        msg - текст с нужными метриками за вчера
        '''

        likes_lastd = df[df['event_date'] == str(date.today() - timedelta(
            days=1))].likes.sum()
        views_lastd = df[df['event_date'] == str(date.today() - timedelta(
            days=1))].views.sum()
        ctr_lastd = likes_lastd / views_lastd
        dau_lastd = df[df['event_date'] == str(date.today() - timedelta(
            days=1))].user_id.count()

        msg = f'Лента новостей (вчера)\n' \
        f'-----------\n' \
        f'CTR : {round(ctr_lastd, 3)}\n' \
        f'DAU : {dau_lastd}\n' \
        f'likes : {likes_lastd}\n' \
        f'views : {views_lastd}'

        return msg

    @task
    def transform_2(df: pd.DataFrame) -> io.BytesIO:
        '''
        Строит графики на основе метрик за неделю
        ----------
        Аргументы:
        df - извлеченный из CH датафрейм
        ----------
        Возвращает:
        plot_object - обьект с изображением(графиками)
        '''

        #Создание нового датафрейма с метриками за неделю
        df['ctr'] = df['likes'] / df['views']
        df_week = df.groupby('event_date').agg({
            'user_id': 'count',
            'likes': 'sum',
            'views': 'sum',
            'ctr': 'mean'
        }).copy()
        df_week.rename(columns={'user_id': 'dau'}, inplace=True)
        df_week.reset_index(inplace=True)
        df_week['event_date'] = df_week['event_date'].dt.date

        #Отрисовка графиков для каждой метрики
        fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 14))
        fig.suptitle('Основные показатели за последние 7 дней', fontsize=25)

        likes = sns.barplot(data=df_week,
                            x='event_date',
                            y='likes',
                            ax=axes[0][0])
        axes[0][0].set_title('Лайки', fontsize=20)
        views = sns.barplot(data=df_week,
                            x='event_date',
                            y='views',
                            ax=axes[0][1])
        axes[0][1].set_title('Просмотры', fontsize=20)
        CTR = sns.barplot(data=df_week, x='event_date', y='ctr', ax=axes[1][0])
        axes[1][0].set_title('CTR', fontsize=20)
        DAU = sns.barplot(data=df_week, x='event_date', y='dau', ax=axes[1][1])
        axes[1][1].set_title('DAU', fontsize=20)

        list_axes = [likes, views, CTR, DAU]

        #Дабавление значений над столбцами диаграмм
        for ax in list_axes:
            for p in ax.patches:
                if ax == CTR:
                    value = '{:.02f}'.format(p.get_height())
                else:
                    value = '{:.0f}'.format(p.get_height())
                ax.annotate(value,
                            (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha='center',
                            va='center',
                            xytext=(0, 10),
                            textcoords='offset points',
                            fontsize=11)

        #Сохранение изолбражения графиков в обьект
        plot_object = io.BytesIO()
        plt.savefig(plot_object)
        plot_object.seek(0)
        plot_object.name = 'metrics_plot.png'
        plt.close()

        return plot_object

    @task
    def send_tg(message: str, plot_object: io.BytesIO, chat_id=None) -> None:
        '''
        Отсылает данные в канал Telegram через бота
        ----------
        Аргументы:
        message - текст с нужными метриками за вчера
        plot_object - обьект с изображением(графиками)
        chat_id - id чата, в который идет отсылка
        '''
        chat_id = chat_id or 220269868

        my_token = '6022521240:AAEs25jL8Wg1cZPb5WpX2884N-w6nyziRow'
        bot = telegram.Bot(token=my_token)

        bot.send_message(chat_id=chat_id, text=message)

        bot.sendPhoto(chat_id=chat_id, photo=plot_object)

    #Вызов нужных функций
    df = extract()

    message = transform_1(df)

    plot_object = transform_2(df)

    send_tg(message=message, plot_object=plot_object, chat_id=-802518328)


#Вызов самого дага
dag_report_krupnov = dag_report_krupnov()
```
