```python
from airflow.decorators import dag, task
from airflow.operators.python import PythonOperator
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
import telegram
import pandahouse as ph
from typing import Tuple


@dag(
    default_args={
        'owner': 's.krupnov',
        'depends_on_past': False,
        'retries': 3,
        'retry_delay': timedelta(minutes=5),
        'start_date': datetime(2023, 3, 28)
    },
    schedule_interval='0 11 * * *',
    catchup=False,
    tags=['example'],
)
def dag_alerts():
    @task
    def extract() -> pd.DataFrame:
        '''
        Выгружает нужные данные из ClickHouse
        ----------
        Возвращает:
        df - датафрейм с нужными данными
        '''
        connection = {
            'host': 'https://clickhouse.lab.karpov.courses/',
            'password': 'dpo_python_2020',
            'user': 'student',
            'database': 'simulator_20230220'
        }

        q_feed = '''
                SELECT toStartOfFifteenMinutes(time) as ts,
                       toDate(ts) as date,
                       formatDateTime(ts, '%R') as hm,
                       uniqExact(user_id) as users_lenta
                FROM simulator.feed_actions
                WHERE ts >=  today() - 1 and ts < toStartOfFifteenMinutes(now())
                GROUP BY ts, date, hm
                ORDER BY ts
                '''

        data = ph.read_clickhouse(query=q_feed, connection=connection)
        return data

    @task
    def check_anomaly(df: pd.DataFrame,
                      metric: str,
                      threshold: float = 0.3) -> Tuple[int, float, float]:
        '''
        Проверяет наличие аномалии в метрике с использованием заданного алгоритма.
        
        Параметры:
            - df: pd.DataFrame
                Датафрейм с данными.
            - metric: str
                Название метрики для проверки на аномалию.
            - threshold: float, optional
                Пороговое значение для определения аномалии. По умолчанию 0.3.
        
        Возвращает:
            Кортеж из трех элементов:
            - int: Состояние предупреждения (1 - есть предупреждение, 0 - нет предупреждения).
            - float: Текущее значение метрики.
            - float: Отклонение от предыдущего значения в процентах.
        '''
        current_ts = df['ts'].max()
        day_ago_ts = current_ts - pd.DateOffset(days=1)

        current_value = df.loc[df['ts'] == current_ts, metric].iloc[0]
        day_ago_value = df.loc[df['ts'] == day_ago_ts, metric].iloc[0]

        if current_value <= day_ago_value:
            diff = abs(current_value / day_ago_value - 1)
        else:
            diff = abs(day_ago_value / current_value - 1)

        is_alert = 1 if diff > threshold else 0

        return is_alert, current_value, diff

    @task
    def run_alerts(is_alert: int,
                   current_value: float,
                   diff: float,
                   data: pd.DataFrame,
                   chat_id: int = 454623234) -> None:
        '''
        Запускает проверку аномалий и отправляет уведомления с графиками в Telegram.
        
        Параметры:
            - is_alert: int
                Состояние предупреждения (1 - есть предупреждение, 0 - нет предупреждения).
            - current_value: float
                Текущее значение метрики.
            - diff: float
                Отклонение от предыдущего значения в процентах.
            - data: pd.DataFrame
                Датафрейм с данными.
            - chat_id: int, optional
                Идентификатор чата в Telegram. По умолчанию 454623234.
        
        Возвращает:
            None
        '''
        if is_alert:
            msg = f'''🎯Метрика: Отправленные сообщения
            Значение на данный момент: {current_value}
            Отклонение от предыдущего значения: +{diff:.2%}

            📊Ссылка на дашборд для детального обзора:
            https://superset.lab.karpov.courses/superset/dashboard/3257/

            📈График:'''

            sns.set(rc={'figure.figsize': (16, 10)})
            plt.tight_layout()

            ax = sns.lineplot(data=data.sort_values(by=['date', 'hm']),
                              x="hm",
                              y="users_lenta",
                              hue="date")

            for ind, label in enumerate(ax.get_xticklabels()):
                if ind % 15 == 0:
                    label.set_visible(True)
                else:
                    label.set_visible(False)
            ax.set(xlabel='time')
            ax.set(ylabel='users_lenta')

            ax.set_title('users_lenta')
            ax.set(ylim=(0, None))

            plot_object = io.BytesIO()
            ax.figure.savefig(plot_object)
            plot_object.seek(0)
            plot_object.name = 'users_lenta.png'
            plt.close()

            bot = telegram.Bot(
                token='1987162789:AAFgHNqBv-v5VXPQcS0btoxtXECUvw8akMs')
            bot.sendMessage(chat_id=chat_id, text=msg)
            bot.sendPhoto(chat_id=chat_id, photo=plot_object)

    data = extract()
    metric = 'users_lenta'
    is_alert, current_value, diff = check_anomaly(data, metric, threshold=0.1)
    run_alerts(is_alert, current_value, diff, data)


dag_alerts = dag_alerts()
```
